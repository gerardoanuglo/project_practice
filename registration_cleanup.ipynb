```python
#import dependencies
import pandas as pd
```

```python
#load csv file into juypter notebook
ca_file = pd.read_csv("/Users/Jerry/Downloads/ca_ev_registrations_public.csv")
ca_file.head()
```

```python
ca_file["Vehicle ID"].drop_duplicates()
```

```python
#remove unneccessary columns
filtered_data = ca_file[["State Abbreviation", "Geography", "Vehicle Name"]]
filtered_data.head()
```

```python
#spilt vehicle name column into make and model
filtered_data[["Make", "Model"]] = filtered_data["Vehicle Name"].str.split(' ', n=1, expand=True)
```

```python
#filter to final columns and change state column name
ca_df= filtered_data[["State Abbreviation", "Make", "Model"]]

ca_df = ca_df.rename(columns={"State Abbreviation":"State"})

#create a count column for total models in the states
a=ca_df.groupby('Model')['Model'].count()
a
```

```python
ca_df["Total Models"]=ca_df.groupby('Make')['Model'].count()
```
